**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python

def prewitt_h(img_in):
    result = numpy.zeros(img_in.shape)
    size = 3
    n = 1
    for p in numpy.ndindex(img_in.shape):
        if p[0]<n or p[1]<n or p[0]>=img_in.shape[0]-n or p[1]>=img_in.shape[1]-n:
            result[p] = 0
        else: 
            upper = img_in[p[0]-n, (p[1]-n):(p[1]+n+1)]
            lower = img_in[p[0]+n, (p[1]-n):(p[1]+n+1)]
            result[p] = (numpy.sum(upper) - numpy.sum(lower))/6
    return result
    
def prewitt_v(img_in):
    result = numpy.zeros(img_in.shape)
    size = 3
    n = 1
    for p in numpy.ndindex(img_in.shape):
        if p[0]<1 or p[1]<1 or p[0]>=img_in.shape[0]-1 or p[1]>=img_in.shape[1]-1:
            result[p] = 0
        else:
            left = img_in[(p[0]-1):(p[0]+2), p[1]-1]
            right = img_in[(p[0]-1):(p[0]+2), p[1]+1]
            result[p] = (numpy.sum(right) - numpy.sum(left))/6
    return result
```

## Edge detection

**<a id="#">Task 4.2.1.</a>** Use your results from **Task 4.1.2 â€“ 4.1.3** to compute the *gradient magnitude* of the image:

```python
def gradientmag(img_in):
    result = numpy.sqrt(numpy.square(prewitt_v(img_in)) + numpy.square(prewitt_h(img_in)))
    return result
```
